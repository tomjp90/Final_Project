```python
# Investigating columns with null values
df.isnull().sum()
```

```python
# Dropping these null values
df = df.dropna() 
df.head()
```

```python
# Checking the length and that removal of na has worked
final_len = len(df)
print(final_len)
print(df.isnull().sum())
```

```python
percent = (final_len/full_len)*100
round(percent,2)
(f"The dataframe has been reduced from {full_len} rows to {final_len} rows, by {round(percent,2)} percent")
```

```python
# Spliiting CouncilArea columns to match crime data LGA format
df["CouncilArea"]= df["CouncilArea"].str.split(" ", n = 1, expand = True) 
df
```

```python
# Splitting the date
df[['Day', 'Month', 'Year']] = df["Date"].str.split("/", expand=True,)
df
```

```python
df.columns
```

```python
# Reducing to columns of interest
# 'Landsize'
df = df.loc[:,['Rooms', 'Type', 'Price', 'Postcode', 'Bathroom', 'Car', 'CouncilArea', 'Month', 'Year']]
df
```
