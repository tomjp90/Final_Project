# Extract - Melbourne House Data 

Not useable because Average of average incidents 

```python
# Imports
import pandas as pd
import numpy as np
```

```python
# Reading in dataframe from CSV
df = pd.read_csv("Resources/Melbourne_housing_FULL.csv")
df.head()
```

# EDA AND TRANSFORM - Melbourne House Data 

```python
full_len = len(df)
full_len
```

```python
# Splitting the date
df[['Day', 'Month', 'Year']] = df["Date"].str.split("/", expand=True,)
df.head()
```

```python
# List of columns
df.columns
```

```python
# Reducing to columns of interest
# 'Landsize'
df = df.loc[:,['Suburb', 'Rooms', 'Type', 'Price', 'Bathroom', 'Car', 'Month', 'Year']]
df.head()
```

```python
# Changing column headers to lower case
df.columns = df.columns.str.lower()
df.head()
```

```python
# Investigating columns with null values
df.isnull().sum()
```

```python
# Dropping these null values
house_df = df.dropna() 
house_df.head()
```

```python
# Checking the length and that removal of na has worked
final_len = len(house_df)
print(final_len)
print(house_df.isnull().sum())
```

```python
difference = full_len - final_len
percent = (difference/full_len)*100
(f"The dataframe has been reduced from {full_len} rows to {final_len} rows, by {round(percent,2)} percent")
```

```python
# Investigating different types of houses
house_df['type'].unique()
```

# Extract - VIC Crime Data

```python
# Reading in dataframe from CSV
crime_vic_df = pd.read_csv("Resources/suburb_vs_crime.csv")
crime_vic_df.head()
```

# EDA AND TRANSFORM - VIC Crime Data

```python
# Capitalising suburb names to perform group by with house_df
crime_vic_df['suburb'] = crime_vic_df['suburb'].str.title()
crime_vic_df.head()
```

```python
# Checking capitalisation
crime_vic_df['suburb'].unique()
```

```python
crime_vic_df.columns
```

```python
# Selecting columns of interest
crime_vic_df = crime_vic_df.loc[:,['suburb', 'average_incident']]
crime_vic_df.head()
```

```python
# Groupby of suburbs 
# ISSUE - I HAVE AVERAGE NUMBER OF INCIDENTS FROM AVERAGE OF EACH TYPE
# SHOULD IT BE TOTAL?
# BEST TO HAVE TOTAL NUMBER OF EACH TYPE AND THEN AVERAGE TOTAL
avg_incidents = crime_vic_df.groupby("suburb")["average_incident"].mean()
crime_df = pd.DataFrame({
    "average_incidents" : avg_incidents
})
crime_df.head()
```

```python
# Checking for null values 
crime_df.isnull().sum()
```

# Transform - merging house and crime data

```python
# Merging two dataframes on suburb
combined_df = pd.merge(house_df, crime_df, how = 'inner', on='suburb')
combined_df.head()
```

```python
# Double checking for null values
combined_df.isnull().sum()
```

```python
combined_len = len(combined_df)
print(f"Combined df length: {combined_len}")
print(f"We have lost {final_len - combined_len} rows of data when merging the 2 dataframes")
```

```python
# Number of suburbs in housing dataframe
house_subs = house_df['suburb'].nunique()
# Number of suburbs in combined dataframe
combined_subs = combined_df['suburb'].nunique()
print(f"Number of suburbs in combined df: {combined_subs}")
print(f'We have lost a total of {house_subs - combined_subs} suburbs from merging housing data with crime data.')
```

```python
# Investigating the loss of  56 suburbs
crimedf_suburbs = crime_vic_df["suburb"].unique()

housedf_suburbs = house_df["suburb"].unique()

combineddf_suburbs = combined_df["suburb"].unique()
```

```python
# Listing the suburbs we have lost through the merge of the 2 dataframes
lost_suburbs = np.setdiff1d(housedf_suburbs, combineddf_suburbs)
print(len(lost_suburbs))
lost_suburbs
```

```python
# Checking that these suburbs never existed in the crime dataframe and therefore haven't lost them during the merge
diff_suburbs = np.setdiff1d(lost_suburbs,crimedf_suburbs)
print(len(diff_suburbs))
diff_suburbs
```
