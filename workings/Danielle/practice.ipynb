```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
X = test_dropped.drop("Price", axis=1)
y = test_dropped['Price'].values.reshape(-1, 1)
```

```python
from sklearn.model_selection import train_test_split

# YOUR CODE HERE
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
```

```python
X_train.head()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python


X_scaler = StandardScaler().fit(X_train)
#y_scaler = StandardScaler().fit(y_train)
```

```python
# taking test train splits and scaling them with the scalers we just created

X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
#y_train_scaled = y_scaler.transform(y_train)
#y_test_scaled = y_scaler.transform(y_test)
```

```python
#model.fit(X_train_scaled, y_train_scaled)
model.fit(X_train_scaled, y_train)
```

```python
#print(f"Training Data Score: {model.score(X_train_scaled, y_train_scaled)}")
#print(f"Testing Data Score: {model.score(X_test_scaled, y_test_scaled)}")

print(f"Training Data Score: {model.score(X_train_scaled, y_train)}")
print(f"Testing Data Score: {model.score(X_test_scaled, y_test)}")
```

```python
predictions = model.predict(X_test_scaled)
predictions
```
